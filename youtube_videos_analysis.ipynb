```python
def main_dataframe_formation(listdir=None, raw_dataframe=None):
    if listdir is None:
        listdir = os.listdir()

    csv_files = [filename for filename in listdir if filename.endswith('.csv')]
    json_files = [filename for filename in listdir if filename.endswith('.json')]

    coded_country_names = [country_name[:2] for country_name in csv_files]
    country_names = [pycountry.countries.get(alpha_2=coded_country_name).name for coded_country_name in
                     coded_country_names]

    if raw_dataframe is None:
        us_csv_filename = csv_files[-1]

        raw_dataframe = pd.read_csv(us_csv_filename, encoding='latin1')

        raw_dataframe['country'] = country_names[-1]

    csv_files = csv_files[:len(csv_files) - 1]
    country_names = country_names[:len(country_names) - 1]

    for file, country_name in zip(csv_files, country_names):
        country_videos = pd.read_csv(file, encoding='latin1')
        country_videos['country'] = country_name
        raw_dataframe = pd.concat([raw_dataframe, country_videos], ignore_index=True)

    categories = {}
    for file in json_files:
        country_category = pd.read_json(file, encoding='latin1')
        for category in country_category.iloc[:, -1]:
            categories.setdefault(category['id'], category['snippet']['title'])

    categories = pd.DataFrame(data=categories.items(),
                              columns=['category_id', 'category']).astype({'category_id': 'int64'})

    dataframe = raw_dataframe.merge(categories, how='inner', on='category_id')

    return dataframe


trending_videos = main_dataframe_formation()
```

Let's build a heat map of the correlation of views/likes/dislikes/comments.

In this case, the correlation is determined by the Pearson correlation coefficient, it shows the linear relation of the variables.

```python
def showing_heatmap_correlation(dataframe, columns):
    corr_matrix = dataframe[columns].corr()
    
    fig, ax = plt.subplots()
    sns.heatmap(corr_matrix,
                annot=True,
                linewidths=0.4,
                cmap=sns.color_palette("coolwarm", 5),
                annot_kws={'fontsize':15})
    
    plt.tick_params(axis='both', which='minor', labelsize=12)
    
    plt.show()
```

```python
showing_heatmap_correlation(dataframe=trending_videos, columns=numerical_columns)
```

Since the correlation between likes and comments is relatively greater than the correlation between dislikes and comments, it is likely that there should be fewer negative comments than positive comments. But this judgment is rather crude, as the nature of the comments is not taken into account. Dislikes and views have the smallest correlation.

Next, since likes and views, dislikes and views have maximum, minimum correlation respectively, let's build a scatter plot to visualize the relationship between these variables.

```python
def showing_scatter_plot(dataframe, x: str, y: str):
    xlabel = x.capitalize()
    ylabel = y.capitalize()
    
    x = dataframe[x]
    y = dataframe[y]
    
    fig, ax = plt.subplots()
    
    plt.scatter(x=x, y=y, color='red', edgecolors='#000000', linewidths=0.5)
    
    ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
showing_scatter_plot(trending_videos, 'views', 'likes')
```

We can see that the likes and views have a positive linear correlation: with more views, more likes.

# Analysis by country

Since videos were entered into this dataset when they hit a trend, there are a large number of repetitions. Moreover, the same video could hit the trends in different countries, and moreover, several times, such a layout significantly distorts the idea of the interests of the residents of each country. Therefore, we can create a DataFrame for further analysis that will consist of videos with a symmetric difference by country. That is, it will contain only those videos that were in the trends in one country only.

To implement such a DataFrame, we can group the main DataFrame by video_id and aggregate by the number of unique countries, eventually selecting only those videos that have no more than one country.

```python
trending_videos.groupby(['video_id']).agg({'country': 'nunique'})
```

```python
trending_videos[trending_videos['video_id'] == '#NAME?']
```

As we can see, some videos do not have their own video_id, so this attribute cannot be taken as a primary key. The problem is that there is no unique value for each video. It can be solved by grouping attributes. To do this, we can take the following set of attributes as the primary key: video_id, publish_date, publish_time.

```python
number_of_countries = trending_videos.groupby(['video_id', 'publish_date', 'publish_time']).agg({'country': 'nunique'}) \
                        .sort_values('country', ascending=False)
```

Note that the number of lines increased, which means that we separated videos that had the same video_id(#NAME?, #VALUE!), to the point that videos could have been published at one point in time, and somehow they did not have their own id. The probability that any of the dataframe rows included information from multiple videos was rushing towards zero.

Now we can created the desired DataFrame.

```python
videos_trending_one_country_0 = number_of_countries[number_of_countries['country'] == 1].index.droplevel([1, 2])
videos_trending_one_country_1 = number_of_countries[number_of_countries['country'] == 1].index.droplevel([0, 2])
videos_trending_one_country_2 = number_of_countries[number_of_countries['country'] == 1].index.droplevel([0, 1])
```

```python
videos_trending_one_country = trending_videos[(trending_videos['video_id'].isin(videos_trending_one_country_0)) &
                                              (trending_videos['publish_date'].isin(videos_trending_one_country_1)) &
                                              (trending_videos['publish_time'].isin(videos_trending_one_country_2))]
```

In this form the information is quite difficult to perceive, so let's build pie charts for each of the countries.

```python
def plotting_pie_plot(dataframe, country: str):
    dataframe = dataframe.sort_values(by=country, axis=1)
    
    labels = dataframe.columns
    sizes = country_category_views_percent.sort_values(by=country, axis=1).loc[country, :].values
    explode = np.zeros_like(labels)
    explode[-1] = 0.05
    redone_labels = [(label if size >= 2 else '') for label, size in zip(labels, sizes)]
    
    
    def my_autopct(pct):
        return ('%1.1f%%' % pct) if pct >= 2 else ''
    
    
    
    fig, ax = plt.subplots(figsize=(10, 10))
    wedges, texts, autotexts  = ax.pie(sizes,
                                       labels=redone_labels,
                                       shadow=True,
                                       explode=explode,
                                       autopct=my_autopct,
                                       startangle=90,
                                       radius=1.3,
                                       colors=sns.color_palette('tab20'))
    
    
    autotexts = [(autotext if size >= 2.5 else '') for autotext, size in zip(autotexts, sizes)]
    
    ax.legend(wedges, labels,
              title="Categories",
              loc="lower right",
              bbox_to_anchor=(1, 0, 0.5, 1))
    
    ax.set_title(f'Pie chart of categories for {country}',
                 fontsize = 15,
                 fontfamily = 'monospace',
                 fontstyle = 'oblique',
                 color='white',
                 backgroundcolor='black',loc='right')
    
    plt.show()
```

```python
country_statistics['likes/dislikes'] = country_statistics.likes / country_statistics.dislikes
country_statistics['views/activity'] = (country_statistics.likes +
                                        country_statistics.dislikes +
                                        country_statistics.comment_count) / country_statistics.views * 100
country_statistics = country_statistics.round(decimals=2)
country_statistics
```

Next, let's look at how the likes and dislikes for each country relate, graphically.

```python
def plotting_likes_dislikes(data):
    
    fig, ax = plt.subplots(figsize=(17,12))
    
    sns.barplot(data=data,
                x='likes',
                y='dislikes')
    
    default_list = data.sort_values('likes')
    
    dislikes_list = default_list['dislikes'].values
    likes_list = default_list['likes'].values
    labels = default_list.index
    
    xlocs = ax.get_xticks()

    for i, y, inf  in zip(range(len(dislikes_list)), dislikes_list, likes_list):
        ax.text(xlocs[i], y, str(inf), color='black', va='bottom', ha='center', rotation=20, fontsize=14)
    
    ax.set_xticklabels(labels, fontsize=11)
    
    plt.show()
    
    
def plotting_ratio_likes_dislikes(data):
    data = data.sort_values(by='likes/dislikes')
    x = np.arange(1, len(data['likes/dislikes']) + 1)
    y = data['likes/dislikes']
    position = x - 1
    
    ax = sns.barplot(x=x, y=y)

    legend_labels  = data.index
    
    for i, y, inf  in zip(x - 1, y, y):
        ax.text(i, y, str(inf), color='black', va='bottom', ha='center', rotation=20, fontsize=14)
        
        
    ax.set_xticks(position)
    ax.set_xticklabels(legend_labels, fontsize=11)
    
    plt.show()
```

The relationship between dislikes and likes.
